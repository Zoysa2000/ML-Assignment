```python
import pandas as pd
df = pd.read_csv('Downloads\healthcare-dataset-stroke-data.csv')
df.head()
```

```python
df_no_id = df.drop('id', axis=1)
```

```python
df_no_id.head()
```

```python
df_no_id['gender'] = df_no_id['gender'].map({'Male': 1, 'Female': 0}).fillna(0).astype(int)
```

```python
df_no_id.head()
```

```python
df_no_id.tail()
```

```python
df_no_id['age'] = df_no_id['age'].astype(int)
```

```python
df_no_id.head()
```

```python
df_no_id
```

```python
average_age = df_no_id.loc[df_no_id['age'] > 18, 'age'].mean()

# Replace ages between 1 and 18 with the calculated average
df_no_id.loc[df_no_id['age'].between(1, 18), 'age'] = average_age

# Convert age column to integer if needed
df_no_id['age'] = df_no_id['age'].astype(int)

# Verify changes
df_no_id.head() 
```

```python
df_no_id.iloc[3089:3099]
```

```python
df_no_id.head()
```

```python
df_no_id['ever_married'] = df_no_id['ever_married'].map({'Yes': 1, 'No': 0})

df_no_id.head()
```

```python
df_no_id['work_type'] = df_no_id['work_type'].map({
    'Private': 1,
    'Govt_job': 2,
    'Self-employed': 3
})
```

```python
most_common_work_type = df_no_id['work_type'].mode()[0]
```

```python
df_no_id['work_type'] = df_no_id['work_type'].replace('children', most_common_work_type)
```

```python
df_no_id['work_type'] = df_no_id['work_type'].fillna(0).astype(int)
```

```python
df_no_id
```

```python
df_no_id.iloc[5030:5035]
```

```python
df_no_id['Residence_type'] = df_no_id['Residence_type'].map({'Urban': 1, 'Rural': 0})
```

```python
df_no_id.head()
```

```python
df_no_id['avg_glucose_level'] = df_no_id['avg_glucose_level'].fillna(0).astype(int)
```

```python
df_no_id.head()
```

```python
# Calculate mean of 'bmi' (excluding NaNs)
mean_bmi = df_no_id['bmi'].mean()

# Replace NaN with mean
df_no_id['bmi'] = df_no_id['bmi'].fillna(mean_bmi)
```

```python
df_no_id['bmi'] = df_no_id['bmi'].astype(int)
```

```python
df_no_id.head()
```

```python
df_no_id.iloc[1860:1869]
```

```python
df_no_id['smoking_status'] = df_no_id['smoking_status'].replace({
    'formerly smoked': 1,
    'never smoked': 2,
    'smokes': 3
})
```

```python
most_common_smoking = df_no_id['smoking_status'].mode()[0]

# Replace 'Unknown' with mode
df_no_id['smoking_status'] = df_no_id['smoking_status'].replace('Unknown', most_common_smoking)

# Optional: convert to int type
df_no_id['smoking_status'] = df_no_id['smoking_status'].astype(int)
```
